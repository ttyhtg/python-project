# 9.2 Python的字符编码

```python
str1 = "我们"
print (str1)
print (type(str1))
```

```python
str1 = "我们"
str_utf8 = str1.encode('utf-8')
print (str_utf8)
print (type(str_utf8))
```

```python
str_decode = str1.encode('utf-8').decode('utf-8')
print (str_decode)
print (type(str_decode))
```

```python
import chardet
str_gbk = "我们".encode('gbk')
chardet.detect(str_gbk)
```

# 9.3 	解决中文编码问题

## 9.3.1 问题1：获取网站的中文显示乱码

```python
import requests
from bs4 import BeautifulSoup

url = 'http://w3school.com.cn/'
r = requests.get(url)
soup = BeautifulSoup(r.text, "lxml")
xx = soup.find('div',id='d1').h2.text
print (xx)
```

```python
r.encoding
```

```python
import requests
from bs4 import BeautifulSoup

url = 'http://w3school.com.cn/'
r = requests.get(url)
r.encoding = 'gb2312'
soup = BeautifulSoup(r.text, "lxml")
xx = soup.find('div',id='d1').h2.text
print (xx)
```

## 9.3.3 问题3：网页使用gzip压缩

```python
import requests
url = 'http://www.sina.com.cn/'
r = requests.get(url)
print (r.text)
```

```python
import chardet
after_gzip = r.content
print ('解压后字符串的编码为',chardet.detect(after_gzip))
print (after_gzip.decode('utf-8'))
```

## 9.3.4 问题4：读写文件的中文乱码

```python
result = open('test_utf8.txt','r').read()
print (result)
```

```python
result_ANSI = open('test_ANSI.txt', 'r', encoding='ANSI').read()
print (result_ANSI)
result_utf8 = open('test_utf8.txt', 'r', encoding='utf-8').read()
print (result_utf8)
```

```python
title = '我们'
with open('title.txt', 'a+', encoding='utf-8') as f:
    f.write(title)
    f.close()
```

```python
import json
title = '我们 love 你们'
with open('title.json','w',encoding = 'utf-8') as f:
    json.dump([title],f)
```

```python
import json
title = '我们 love 你们'
with open('title.json','w',encoding = 'utf-8') as f:
    json.dump([title],f,ensure_ascii=False)
```
